```python
investment = pd.read_csv("Media Investment.csv")
```

```python
investment.head()
```

```python
investment.replace(np.nan,0,inplace = True)
investment.head()
```

```python
investment['Date'] = pd.to_datetime(investment[['Year', 'Month']].assign(DAY=1))
investment.set_index('Date', inplace = True)
investment.dtypes
```

```python
nos_week = [['7', 4], ['8', 5], ['9', 4], ['10', 4], ['11', 5], ['12', 4], ['1', 4], ['2', 5], ['3', 4], ['4', 4], ['5', 5], ['6', 4]]
```

```python
number_weeks = pd.DataFrame(nos_week, columns = ['Month', 'week_nos'])
number_weeks['Month'] = number_weeks['Month'].astype('int')
```

```python
number_weeks.dtypes
```

```python
investment = investment.merge(number_weeks, how  = "left", on=["Month"])
```

```python
investment.head(12)
```

```python
investment['week_nos'] = investment['week_nos'].astype('float')
investment.dtypes
```

```python
investment['Total Investment'] = investment['Total Investment'] / investment['week_nos'] 
investment['TV'] = investment['TV'] / investment['week_nos'] 
investment['Digital'] = investment['TV'] / investment['week_nos'] 
investment['Sponsorship'] = investment['TV'] / investment['week_nos'] 
investment['Content Marketing'] = investment['TV'] / investment['week_nos'] 
investment['Online marketing'] = investment['TV'] / investment['week_nos'] 
investment['Affiliates'] = investment['TV'] / investment['week_nos'] 
investment['SEM'] = investment['TV'] / investment['week_nos'] 
investment['Radio'] = investment['TV'] / investment['week_nos'] 
investment['Other'] = investment['TV'] / investment['week_nos'] 
```

```python
dataset_merged = dataset.merge(investment, how = "left", on= ["Month"])
```
